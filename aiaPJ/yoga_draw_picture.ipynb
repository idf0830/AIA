```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
y_pred_single = pd.read_csv('X1_單價_Y_pred.csv')
y_test_single = pd.read_csv('X1_單價_Y_test.csv')
y_pred_sum = pd.read_csv('X1_總價_Y_pred.csv')
y_test_sum = pd.read_csv('X1_總價_Y_test.csv')

y_pred_single = y_pred_single['0']
y_test_single = y_test_single['0']
y_pred_sum = y_pred_sum['0']
y_test_sum = y_test_sum['0']
```

## Draw Picture
- plot1_single

```python
# 真实值与预测值的关系# 设置绘图风格
fig, ax = plt.subplots(figsize = (10, 10))
plt.style.use('ggplot')
# 设置中文编码和负号的正常显示
plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'

# 散点图
plt.scatter(y_test_single, y_pred_single, label = 'data', c = 'blue')
# 回归线
plt.plot([y_test_single.min(), y_test_single.max()], [y_pred_single.min(), y_pred_single.max()], 'r--', lw=2, label = 'fit line')

# 添加轴标签和标题
plt.title('real_label VS. prediction')
plt.xlabel('real_label')
plt.ylabel('prediction')

# 去除图边框的顶部刻度和右边刻度
plt.tick_params(top = 'off', right = 'off')
# 添加图例
plt.legend(loc = 'upper left')
# 图形展现
plt.savefig('plot1_single.png')
plt.show()
```

- plot2_single

```python
# ====== 圖示法完成方差齊性的判斷 ======
# 標準化殘差與預測值之間的散點圖
fig, ax = plt.subplots(figsize = (10, 10))
plt.scatter(y_pred_single,y_test_single - y_pred_single,c = 'blue')
plt.xlabel('fitted value')
plt.ylabel('residual')

# 添加水平參考線
plt.axhline(y = 0, color = 'r', linewidth = 2)
plt.savefig('plot2_single.png')
plt.show()
```
